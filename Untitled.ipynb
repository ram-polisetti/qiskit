```python
import qiskit
print(qiskit.__version__)
```

```python
from qiskit import *
```

```python
qc = QuantumCircuit(2, name='qc')
```

```python
qc.x(0)
```

```python
%matplotlib inline
qc.draw(output='mpl')
```

```python
qc.cx(1, 0)
```

```python
%matplotlib inline
qc.draw(output='mpl')
```

```python
qc.measure_all()
```

```python
qc.draw(output='mpl')
```

```python
from qiskit_aer import Aer
```

```python
Aer.backends()
```

```python
emulator = Aer.get_backend('aer_simulator')
```

```python
job = emulator.run(qc, shots=8024)
```

```python
hist = job.result().get_counts(qc)
```

```python
from qiskit.visualization import plot_histogram
```

```python
plot_histogram(hist)
```

```python
state_vector = Aer.get_backend('statevector_simulator').run(qc).result().get_statevector()
print(state_vector)
```

```python
from qiskit.visualization import plot_bloch_vector
from matplotlib import pyplot as plt
```

```python
plot_bloch_vector(state_vector, title='Bloch Vector')
```
