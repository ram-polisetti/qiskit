```python
from qiskit import *
from qiskit_aer import Aer
from qiskit.visualization import plot_bloch_multivector, visualize_transition
```

```python
qc = QuantumCircuit(1, 1)
qc.h(0)
qc.measure(0, 0)
%matplotlib inline
qc.draw(output='mpl')
```

the qubit is in superposition after the hadamard gate but when we measure it if will fall into either ket+ or ket-

```python
emulator = Aer.get_backend('statevector_simulator')
result = emulator.run(qc).result()
statevector = result.get_statevector()
print(statevector)
plot_bloch_multivector(statevector)
```
